create yaml file as instruction

```python
yaml_content = """
path: /root/.cache/kagglehub/datasets/nadinpethiyagoda/vehicle-dataset-for-yolo/versions/1
train: images/train
val: images/valid

nc: 6
names: ['car', 'motorbike', 'threewheel', 'van', 'bus', 'truck']
"""

with open("/root/.cache/kagglehub/datasets/nadinpethiyagoda/vehicle-dataset-for-yolo/versions/1/vehicle_data.yaml", "w") as f:
    f.write(yaml_content)

print("✅ vehicle_data.yaml created!")
```
